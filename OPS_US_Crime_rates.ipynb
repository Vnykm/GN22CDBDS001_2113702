### Step 3. Assign it to a variable called crime.

```python
crime = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/US_Crime_Rates/US_Crime_Rates_1960_2014.csv')
crime.head()
```

### Step 4. What is the type of the columns?

```python
crime.dtypes
```

##### Have you noticed that the type of Year is int64. But pandas has a different type to work with Time Series. Let's see it now.

### Step 5. Convert the type of the column Year to datetime64

```python
crime['Year'] = pd.to_datetime(crime['Year'], format='%Y')
crime.dtypes
```

```python
crime.head()
```

### Step 6. Set the Year column as the index of the dataframe

```python
crime.set_index('Year', inplace=True)
crime.head()
```

### Step 7. Delete the Total column

```python
del crime['Total']
crime.head()
```

### Step 8. Group the year by decades and sum the values

#### Pay attention to the Population column number, summing this column is a mistake

```python
crime.groupby((crime.index.year//10)*10).sum()
```

Since, population should be average of the all the years in Decade:

```python
## This gives the average popluation for the decade with the sum of crimes reported in each decade
df = crime.groupby((crime.index.year//10)*10).sum()
df['Population'] = crime.groupby((crime.index.year//10)*10)['Population'].mean().apply(round)
df.head()
```

### Step 9. What is the most dangerous decade to live in the US?

```python
crimes_reported = df.iloc[:, 2:].sum(axis = 1)
df['crime_ratio'] = (crimes_reported/df['Population'])*100
df
```

To determine most dangerous decade crime ratio can be calculated as ((total crimes reported)/population)*100 which gives crime per 100 people 

```python
df.sort_values(by=['crime_ratio'], ascending=False).head(1)
```

Hence 1980 was the most dangerous decade to live in the US
