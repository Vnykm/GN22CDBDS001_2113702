Importing Libraries


```python
import pandas as pd
import numpy as np
```

### Getting the dataset

```python
df = pd.read_csv('https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/04_Apply/Students_Alcohol_Consumption/student-mat.csv')
df.head()
```

### slice the dataframe from 'school' until the 'guardian' column

```python
stu_gua = df.loc[:,'school':'guardian']
stu_gua.head()
```

### lambda function that will capitalize strings.

```python
cap = lambda S : S.capitalize()
```

### Capitalize both Mjob and Fjob

```python
stu_gua['Mjob'].apply(cap)
stu_gua['Fjob'].apply(cap)
```

### Print the last elements of the data set.

```python
stu_gua.tail(10)
```

###  Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

```python
stu_gua['Mjob'] = stu_gua['Mjob'].apply(cap)
stu_gua['Fjob'] = stu_gua['Fjob'].apply(cap)
stu_gua.tail(10)
```

### Function called majority that returns a boolean value to a new column called legal_drinker

```python
def majority(age):
  if age>17:
    return True
  else:
    return False
```

```python
  stu_gua['legal_drinker'] = stu_gua['age'].apply(majority)
  stu_gua.head()
```
