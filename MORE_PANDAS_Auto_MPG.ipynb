# MPG Cars

### Introduction:
The following exercise utilizes data from [UC Irvine Machine Learning Repository](https://archive.ics.uci.edu/ml/datasets/Auto+MPG)

### Importing the necessary libraries

```python
import pandas as pd
import numpy as np
```

   ### Importing data and Assigning each to a variable called cars1 and cars2

```python
cars1 = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/cars1.csv')
cars2 = pd.read_csv('https://raw.githubusercontent.com/a-forty-two/COG_GN22CDBDS001_MARCH_22/main/cars2.csv')
```

### Checking Data 

```python
cars1.head()
```

```python
cars2.head()
```

### Droping NULL columns from the dataset

```python
cars1.dropna(axis=1, inplace=True)
cars2.dropna(axis=1, inplace=True)
```

```python
cars1.head()
```

```python
cars2.head()
```

Number of observations in each dataset

```python
print(cars1.shape)
print(cars2.shape)
```

cars1 has 198 and cars2 has 200 observations

### Joining cars1 and cars2 into a single DataFrame called cars

```python
cars = pd.concat([cars1, cars2])
cars.head()
```

```python
cars.shape
```

Cars have 398 (198+200) observations

### Creating a random number Series from 15,000 to 73,000.

```python
owners = np.random.randint(15000, 73000, cars.shape[0])
```

### Step 8. Adding the column owners to cars

```python
cars['owners'] = owners
```
